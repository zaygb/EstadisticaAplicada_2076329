```python
# Librerias básicas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Librerias especializadas 
from lifelines import CoxPHFitter, KaplanMeierFitter, ExponentialFitter, WeibullFitter, LogNormalAFTFitter
import seaborn as sns
```

```python
# Data
df_cancer = pd.read_csv("https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/cancer.csv")
```

```python
df_cancer.head()
```

Los datos constan de 228 observaciones y 10 variables/columnas.
La descripcion de las variables es la siguiente: 

**time (d1)**: tiempo de supervivencia en días\
**status (d2)**: estado de censura 1 = censurado, 2 = muerto\
**age (i1)**: Edad en años\
**sex (i2)**: Masculino = 1 Femenino = 2\
**ph.ecog (i3)**: puntuación de rendimiento ECOG según la calificación del médico. 0 = asintomático, 1 = sintomático pero completamente ambulatorio, 2 = en cama <50% del día, 3 = en cama > 50% del día pero no encamado, 4 = encamado\
**ph.karno (i4)**: puntuación de desempeño de Karnofsky (mala = 0; buena = 100) calificada por el médico\
**pat.karno (i4)**: puntuación de rendimiento de Karnofsky según la calificación del paciente\
**meal.cal (i5)**: Calorías consumidas en las comidas\
**wt.loss (i6)**: Pérdida de peso en los últimos seis meses.

```python
# Estadísticas descriptivas -- ajuste datos
## Hacer columnas binarias (0 y 1)
df_cancer['status'] = df_cancer['status'] - 1
df_cancer['sex'] = df_cancer['sex'] - 1
## status 1 = muerto y 0 = censurado
# sex 1 = femenino y 0 = masculino
```

```python
# Eliminar variable "inst"
df_cancer = df_cancer.drop('inst', axis=1)
```

```python
# Verificar que sean numericas para poder utilizarlas
df_cancer.dtypes
```

```python
# Verificar missing values
df_cancer.isnull().sum()
```

```python
##eliminar / imputar datos
# imputar (asignar) los valores perdidos
df_cancer['ph.karno'] = df_cancer['ph.karno'].fillna(df_cancer['ph.karno'].mean())
df_cancer['pat.karno'] = df_cancer['pat.karno'].fillna(df_cancer['pat.karno'].mean())
df_cancer['meal.cal'] = df_cancer['meal.cal'].fillna(df_cancer['meal.cal'].mean())
df_cancer['wt.loss'] = df_cancer['wt.loss'].fillna(df_cancer['wt.loss'].mean())


```

```python
# eliminar los que tienen valores perdidos
df_cancer = df_cancer.dropna()
```

```python
df_cancer['ph.ecog'] = df_cancer['ph.ecog'].astype('int64')
```

```python
### lifelines convertir en dummys las variabales categoricas
dummies = pd.get_dummies(df_cancer['ph.ecog'], prefix='ecog').astype(int)
```

```python
dummies = dummies[['ecog_1', 'ecog_2']]
df_cancer = pd.concat([df_cancer, dummies], axis=1)
df_cancer = df_cancer.drop('ph.ecog', axis=1)
```

```python
# estadisticas descriptivas
df_cancer.describe()
```

El 72% de los individuos fallecieron, el 39% son mujeres, la edad promedio de 62

```python
## distribucion de T
plt.hist(df_cancer['time'], bins=30)
plt.show()
```

```python
#Histograma
sns.histplot(df_cancer['time'], bins=30, kde=True)
## añadir curva
sns.kdeplot(df_cancer['time'])
plt.xlim(0,1000)
plt.title('Histograma del tiempo de Supervivencia')
plt.xlabel('días')
plt.ylabel('frecuencia')
plt.show()
```

```python
## Estimasción del modelo
cph1 = CoxPHFitter().fit(df_cancer,'time','status')
cph1.print_summary()
```

**age** su p-value = 0.17 > 0.05 = alfa, entonces no rechazamos Ho, es decir que la edad no tiene efecto en el riesgo de morir de cáncer

**sex** su p-value es menor a 0.005 , entonces rechazamos Ho, es decir, que ser mujer disminuye el riesgo de morir de cáncer en 44% en comparación con los hombres

**ph.karno, pat.karno, meal.cal, wt.loss** No rechazamos Ho, significa que las variables no tienen efecto en el riesgo de morir por cáncer

**ecog_1 = ambulatorio** Rechazamos Ho, significa que los ambulatorios tiene mayor riesgo de morir de cancer, tiene 74% más riesgo de morir de cáncer, que un asintomatico de cancer

**ecog_2 = cama** Rechazamos Ho, significa que los que estan en cama tiene mayor riesgo de morir de cancer, tiene 194% más riesgo de morir en comparación con un asintomático

Esta base es de EU en los 80's, 

arrest 1 = volvio a ser arrestado, 0 = censurado

week = semanas que tardo en volver a ser arrestado


```python
## Vamos a estimar un modelo de cox
cph2 = CoxPHFitter().fit(rossi, 'week', 'arrest')
cph2.print_summary()
```

**fin** No rechazamos Ho, es decir, darles dinero cuando los liberan no tiee efecto sobre el riesgo de volver a ser arrestado

**age** Rechazamos Ho, la edad disminuye el riesgo de volver a ser arrestado, por cada año que se cumple el riesgo de volver a ser arrestado disminuye en 6%

**race** No rechazamos Ho, ser negro no afecta el riesgo de volver a ser arrestado en comparación con los caucasicos

**wexp, mar, paro** No rechazamo Ho, las vriables no afectan el riesgo de volver a ser arrestado

**prio** Rechazamos Ho, que el haber estado varias veces en la carcel aumenta el riesgo de ser arrestado otra vez, por cada vez que estuvo en la carcel su riesgo de volver aumenta en 10%

el modelo de cox se basa en que el riesgo es constante en el tiempo (riesgo proporcional)

```python
cph2.check_assumptions(rossi, p_value_threshold=0.05)
## Ho viola el supuesto
## Ha no se viola el supuesto
```
