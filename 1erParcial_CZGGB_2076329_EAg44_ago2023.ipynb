```python
#Librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Datos
peliculas = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/peliculas.csv')
poblacion_Asgard = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/poblacion_Asgard.csv')
```

## 1 PELICULAS

```python
## a)
num_pelis = peliculas.groupby('title')['title'].count().count()
num_cat = peliculas.groupby('category')['category'].count().count()
mean_length = peliculas['length'].mean()
```

```python

print('Hay', num_pelis, 'peliculas,', num_cat, 'categorias y en promedio duran', mean_length, 'minutos')
```

```python
## b)
peliculas_dormir = peliculas.query('category=="Documentary"').query('length>120')
peliculas_dormir
```

```python
## c
peliculas.groupby('category')['category'].count()
```

```python
#d 
peliculas.groupby('category')['length'].mean().round(2)
```

```python
#e
peliculas['type'] = np.where(peliculas['length'] < 90, 'Corta', np.where(peliculas['length'] > 150, 'Larga', 'Normal'))
peliculas
```

## 2 POBLACION ASGARD

```python
## a
poblacion_Asgard.sort_values(['Estado', 'Municipio'])
```

```python
## b
num_Estados = poblacion_Asgard.groupby('Estado')['Estado'].count().count()
num_Municipios = poblacion_Asgard.groupby('Municipio')['Municipio'].count().count()
print('En Asgard hay', num_Estados, 'Estados y', num_Municipios, 'Municipios')
```

```python
#Municipios por estado
poblacion_Asgard.groupby('Estado')['Municipio'].count()
```

```python
# c Poblacion Total
print('Poblacion Total de Asgard:', poblacion_Asgard['Población'].sum())
```

```python
#Poblacion por estado
poblacion_Asgard.groupby('Estado')['Población'].sum()
```

```python
# d
mediana = poblacion_Asgard['Población'].median()
poblacion_Asgard['Clasificación'] = np.where(poblacion_Asgard['Población'] < mediana, 'Poco poblados', 'Muy poblados')
poblacion_Asgard
```

```python
#e 5 más poblados
poblacion_Asgard.sort_values('Población', ascending = False).head(5)
```

```python
# 5 menos poblados
poblacion_Asgard.sort_values('Población', ascending = True).head(5)
```

```python
# f
max_id = poblacion_Asgard.groupby('Estado')['Población'].idxmax()
poblacion_Asgard.loc[max_id]
```

## 3. a ) Los 15 municipios más poblados de Asgard

```python
poblacion_Asgard.set_index('Municipio').sort_values('Población', ascending = False).head(15).plot.bar()
plt.title('15 Municipios más poblados de Asgard')
plt.xlabel('Municipio')
plt.ylabel('Habitantes')
plt.show()
```

## b ) Cuantas peliculas hay por genero

```python
peliculas.set_index('category').groupby('category')['title'].count().sort_values(ascending = False).plot.bar()
plt.title('Peliculas por género')
plt.xlabel('género')
plt.ylabel('número de peliculas')
plt.show()
```
